```python
train['CryoSleep'].fillna(False, inplace=True)
train['Cabin'].fillna('0/0/0', inplace=True)
train['HomePlanet'].fillna(train.HomePlanet.mode()[0], inplace=True)
train['Destination'].fillna(train.Destination.mode()[0], inplace=True)
train['VIP'].fillna(train.VIP.mode()[0], inplace=True)
train['Age'].fillna(train.Age.mean(), inplace=True)
train[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']] = train[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].fillna(0)


test['CryoSleep'].fillna(False, inplace=True)
test['Cabin'].fillna('0/0/0', inplace=True)
test['HomePlanet'].fillna(test.HomePlanet.mode()[0], inplace=True)
test['Destination'].fillna(test.Destination.mode()[0], inplace=True)
test['VIP'].fillna(test.VIP.mode()[0], inplace=True)
test['Age'].fillna(test.Age.mean(), inplace=True)
test[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']]= test[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].fillna(0)
```

```python
#Deck and Side possibly useful, Cabin and Cabin Num useless based on EDA from prior notebook
#Note: well... i kinda reset the other notebook to a single blank cell by accident...
test[['Deck', 'Num', 'Side']] = test['Cabin'].str.split('/', expand=True)
train[['Deck', 'Num', 'Side']] = train['Cabin'].str.split('/', expand=True)
test.drop(['Cabin', 'Num'] , axis = 1, inplace = True)
train.drop(['Cabin', 'Num'] , axis = 1, inplace = True)
```

```python
#grouping Age into groups so it's more useful
train['AgeGroup'] = 7
for i in range(6):
    train.loc[(train.Age >= 10*i) & (train.Age < 10*(i + 1)), 'AgeGroup'] = i
test['AgeGroup'] = 7
for i in range(6):
    test.loc[(test.Age >= 10*i) & (test.Age < 10*(i + 1)), 'AgeGroup'] = i
```
